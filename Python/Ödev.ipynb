Yukarı yazdığımız "Araç Kiralama" projesine çok benzer bir proje yazmanızı istiyoruz. Yazacağımız projenin taslağını aşağıda veriyorum.
Araçkirala ile çok benzer özellikler gösteren bir Magaza superclass'ımız olacak. Bu super classımız altında aşağıda notları bulunan fonksiyonlar çalışacak.
Super classımızın child classları olarak televizyon ve telefon olacak. Yine araç kiralamaya çok benzer olarak sadece televizyon classına bir gümrük vergisi metodu uygulayacağız ve sizin belirlediğiniz bir gümrük vergisi satın alma işleminden sonra fatura üzerinde fatura tutarına eklenecek.
Son olarak bir müşteri classı yazacağız ve müşteri bize stok ve fiyat sorgulama ile satın alma taleplerinde bulunabilecek.

```python
class Magaza():
    
    def __init__(self,stok):
        self.stok = stok
        
    def ürünAdı(self):
        print(f"Ürünün adı {self.name}'dur.")
        return self.name
        
    def stokSorgula(self):
        print(f"Alınabilir {self.name} sayısı {self.stok} kadardır.")
        return self.stok
    
    def fiyatSorgula(self):
        if self.name == "Telefon":
            print(f"{self.name}'un fiyatı 2500'dür.")
        
        else:
            print(f"{self.name}'un fiyatı 5700'dür.")
        
    
    def ürünSatınal(self,n):
            if(self.stok <= 0):
                print("Telefon stoğu tükenmiştir.")
            else:
                self.stok -= n      
            
class Telefon(Magaza):
    def __init__(self,stok):
        self.name = "Telefon"
        super().__init__(stok)
        
class Televizyon(Magaza):
    global vergi_oranı
    vergi_oranı = 0.01
    
    def __init__(self,stok):
        self.name = "Televizyon"
        super().__init__(stok)
        
    def gümrükVergisi(self, fatura):
        fatura = fatura + (fatura * vergi_oranı)
        return fatura
    
class Müşteri():
    def __init__(self):
        self.televizyon_adet = 40
        self.telefon_adet = 80
        self.fatura = 0

    
    def fiyatSorgula(self, tip):
        if tip == "Telefon":
            temp = Telefon(self.telefon_adet)
            temp.fiyatSorgula()
        elif tip == "Televizyon":
            temp = Televizyon(self.televizyon_adet)
            temp.fiyatSorgula()
        else:
            print("Sadece Telefon ve Televizyon bulunmaktadır.")
    
    def stokSorgula(self,tip):
        if tip == "Telefon":
            temp = Telefon(self.telefon_adet)
            temp.stokSorgula()
        elif tip == "Televizyon":
            temp = Televizyon(self.televizyon_adet)
            temp.stokSorgula()
        else:
            print("Sadece Telefon ve Televizyon bulunmaktadır.")        
    
    def ürünSatınal(self, n, tip):
        if tip == "Telefon":
            if(self.telefon_adet <= 0):
                print("Telefon stoğu tükenmiştir.")
            else:
                self.telefon_adet -= n
                self.fatura += 2500*n
                print(f"Güncel faturanız: {self.fatura}")
                
        elif tip == "Televizyon":
            if(self.televizyon_adet <= 0):
                print("Televizyon stoğu tükenmiştir.")
            else:
                temp = Televizyon(self.televizyon_adet)
                self.televizyon_adet -= n
                fiyat = 5700*n
                self.fatura = temp.gümrükVergisi(fiyat)
                print(f"Güncel faturanız: {self.fatura}")
        else:
            print("Sadece Telefon ve Televizyon bulunmaktadır.")      
            
    
```
